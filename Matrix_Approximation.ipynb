```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
import os
plt.style.use('seaborn-darkgrid')
```

```python
A = imread('images/tyrion.jpg') #imported the jpg
print(A.shape)
plt.imshow(A)
plt.axis('off') #x-y axis disabled
plt.show()
```

```python
X = np.mean(A,-1) #convert it into gray-scale
plt.imshow(X).set_cmap('gray') #convert it into gray-scale
plt.axis('off')
plt.show()
```

```python
#full_matrices = False shows that we want to do the economy SVD
U, S, VT = np.linalg.svd(X, full_matrices = False) 
print(U.shape, S.shape, VT.shape)

S = np.diag(S) # S was an array of eigenvalues it is nw become a diagonal matrix 

j = 1 # Numbering the figure
for r in (5,20,100):
  # Matrix Approsimation r is the rank
  Xapprox = U[:,:r] @ S[:r,:r] @ VT[:r,:] # S[:r1,:r2] means r1 rows of S and r2 cols of S
  plt.figure(j)
  j += 1
  plt.imshow(Xapprox).set_cmap('gray')
  plt.axis('off')
  plt.title('r = ' + str(r))
  plt.show()
```

```python
plt.figure(1)
plt.semilogy(S)# this is log(σj) vs j
plt.title('Singular Values')
plt.show()
```

The following graph is the cumulative sum which shows the fraction of energy captured by the first $r$ singular values. 
$$\begin{equation*}\displaystyle\sum_{i=1}^j \sigma_i \over \displaystyle \sum_{i=1}^m \sigma_i \end{equation*}  $$

```python
plt.figure(2)
plt.plot(np.cumsum(S)/np.sum(S))
plt.title('Singular Values : Cumulative Sum')
plt.show()
```
